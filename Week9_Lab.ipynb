**1.	Merge Dataframes**

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
print("Dataframe 1:\n",df1)


df2 = pd.DataFrame({ "city": ["chicago","new york","orlando"], "humidity": [65,68,75], }) 
print("\nDataframe 2:\n",df2)

df3 = pd.merge(df1, df2, on="city")
print("\nDataframe after merging based on city:\n",df3)

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
print("\nUpdated Dataframe 1:\n",df1)

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
print("\nUpdated Dataframe 2:\n",df2)

df3=pd.merge(df1,df2,on="city",how="inner")
print("\nInner merge:\n",df3)

df3=pd.merge(df1,df2,on="city",how="outer")
print("\nOuter merge:\n",df3)

df3=pd.merge(df1,df2,on="city",how="left")
print("\nLeft merge:\n",df3)


df3=pd.merge(df1,df2,on="city",how="right") 
print("\nRight merge:\n",df3)

df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1


df2 = pd.DataFrame({ "city": ["chicago","new york","san diego"], "temperature": [21,14,35], "humidity": [65,68,71] }) 
df2

df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3


df1 = pd.DataFrame({ "city": ["new york","chicago","orlando"], "temperature": [21,14,35], }) 
df1.set_index('city',inplace=True) 
df1

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2


df1.join(df2,lsuffix='_l', rsuffix='_r')
```

**5.	Crosstab**

```python
%cd /content
%cd drive/My Drive/Colab Notebooks
import pandas as pd
df = pd.read_excel("survey.xlsx") 
df

print(pd.crosstab(df.Nationality,df.Handedness))

pd.crosstab(df.Sex,df.Handedness)

pd.crosstab(df.Sex,df.Handedness, margins=True)

print(pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True))
# marigins: get summation

print(pd.crosstab([df.Nationality, df.Sex], [df.Handedness], margins=True))

print(pd.crosstab(df.Sex, df.Handedness, normalize='index'))

import numpy as np 
print(pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average))
```
