**5.	Stack Plots**

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='sleeping', linewidth=5)
plt.plot([],[],color='c', label='eating', linewidth=5)
plt.plot([],[],color='r', label='working', linewidth=5)
plt.plot([],[],color='k', label='playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')	
plt.legend()
plt.show()
```

**6.	Pie Charts**

```python
import matplotlib.pyplot as plt
slices = [7,2,2,13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c','m','r','b']

plt.pie(slices, labels = activities, colors=cols, startangle=90, 
shadow = True, explode=[0, 0.1,0,0], autopct='%1.1f%%')
# explode - pull out;	autopct - change to percentage 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')	
plt.legend()
plt.show()
```

**Lab Assignment:
3.The shape of a limacon can be defined parametrically as 
r = r_0  +cos, 
x = r cos, 
y = r sin  
When r_0=1, this curve is called a cardioid. Use this definition to plot the shape of a limacon for r_0=0.8, r_0=1.0, and r_0=1.2. Be sure to use enough points that the curve is closed and appears smooth (except for the cusp in the cardioid). Use a legend to identify which curve is which. Save the plot as a .pdf file and send me a copy with your work.**



```python
import numpy as np
from matplotlib import pyplot as plt

theta = np.linspace(0, 10, 200)

r = 0.8 + np.cos(theta) # r = r0 + cos 
r1 = 1.0 + np.cos(theta)
r2 = 1.2 + np.cos(theta)

# x and y values for r=0.8
x = r * np.cos(theta)
y = r * np.sin(theta)

#x and y values for r = 1.0
x1 = r1 * np.cos(theta)
y1 = r1 * np.sin(theta)

#x and y values for r = 1.2
x2 = r2 * np.cos(theta)
y2 = r2 * np.sin(theta)

print("\nplot licamon for r = 0.8")
plt.plot(x,y,color='b')
plt.title('r=0.8')
plt.show()

print("\nplot licamon for r = 1.0")
plt.plot(x1,y1,color='r')
plt.title('r=1.0')
plt.show()

print("\nplot licamon for r = 1.2")
plt.plot(x2,y2,color='black')
plt.title('r=1.2')
plt.show()
```
